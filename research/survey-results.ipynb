```python

tutor_results = [9, 7, 9, 10, 8]
lecturer_results = [10, 2]
coordinator_results = [7]


plt.bar([1,2,3,4,5], tutor_results, label="Tutor")
plt.bar([6], coordinator_results, label="Team-Project Co-ordinator")
plt.bar([7,8], lecturer_results, label="Lecturer")

plt.xlabel("Participant Number")
plt.ylabel("Perceived Valuability (1-10)")

plt.legend(loc="upper right", bbox_to_anchor=(1.52, 0.8))

plt.savefig("figure.png")
plt.show()
```
